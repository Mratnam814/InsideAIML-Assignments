```python
def convert_to_list(n):
  a=[]
  for i in n:
    a.append(i)
  n=a
  return n


```

```python
Current_Rank=convert_to_list(df['Current Rank'])
Year=convert_to_list(df['Year'])
earnings=convert_to_list(df['earnings ($ million)'])
Sport=convert_to_list(df['Sport'])
Nationality=convert_to_list(df['Nationality'])
PYrank=convert_to_list(df['Previous Year Rank'])
```

```python
#LinePlot
plt.plot(Year,earnings,color="g",linestyle="-",linewidth=1)
plt.title("LINE PLOT")
plt.xlabel("Earnings")
plt.ylabel("Year")

plt.grid(True)
plt.show()
```

```python
def splitter(n):
  Top_Nations=[]
  Top_Nations_Num=[]
  Finest_Performers=[]
  Finest_Performers_Num=[]
  Good_Performers=[]
  Good_Performers_Num=[]
  Poor_show=[]
  Poor_show_Num=[]
  l=pd.DataFrame(n.value_counts())
  l=l.to_dict()
  for i in l.values():
    j=i
  N_Name=[]
  N_no=[]
  for i in j.values():
    N_no.append(i)
  for i in j.keys():
    N_Name.append(i)
  for j in range(0,5):
    Top_Nations.append(N_Name[j])
    Top_Nations_Num.append(N_no[j])
  for j in range(5,10):
    Finest_Performers.append(N_Name[j])
    Finest_Performers_Num.append(N_no[j])
  for j in range(10,13):
    Good_Performers.append(N_Name[j])
    Good_Performers_Num.append(N_no[j])
  for j in range(13,22):
    Poor_show.append(N_Name[j])
    Poor_show_Num.append(N_no[j])
  return Top_Nations,Top_Nations_Num,Finest_Performers,Finest_Performers_Num,Good_Performers,Good_Performers_Num,Poor_show,Poor_show_Num





n=df['Nationality']
splitter(n)
Top_Nations = splitter(n)[0]
Top_Nations_Num = splitter(n)[1]
Finest_Performers = splitter(n)[2]
Finest_Performers_Num=splitter(n)[3]
Good_Performers=splitter(n)[4]
Good_Performers_Num=splitter(n)[5]
Poor_show=splitter(n)[6]
Poor_show_Num=splitter(n)[7]
n=df['Sport']
splitter(n)
Top_sport = splitter(n)[0]
Top_sport_num=splitter(n)[1]
Finest_sport=splitter(n)[2]
Finest_sport_num=splitter(n)[3]
Good_sport=splitter(n)[4]
Good_sport_num=splitter(n)[5]
Low_played=splitter(n)[6]
Low_played_num=splitter(n)[7]
n=df[ 'earnings ($ million)']
splitter(n)
Top_earnings=splitter(n)[0]
Top_earnings



```

```python
#LinePlot
plt.plot(Top_Nations,Top_Nations_Num,color="g",linestyle="-",linewidth=1)
plt.title("LINE PLOT")
plt.xlabel("Top_Nations")
plt.ylabel("Rankers in Forbes")

plt.grid(True)
plt.show()
```

```python
df['Nationality'].groupby(df['earnings ($ million)'])
```

```python
df1=df.groupby(['Nationality']).agg({'earnings ($ million)':'mean'})
```

```python
df1=pd.DataFrame(df1)
```

```python
df1=df1.to_dict()
```

```python
for i in df1.values():
  j=i
keys=[]
values=[]
keys_top=[]
Values_top=[]
keys_middle=[]
Values_middle=[]
keys_bottom=[]
Values_bottom=[]

for i in j:
  keys.append(i)
  values.append(j[i])
for i in range(0,5):
  keys_top.append(keys[i])
  Values_top.append(values[i])
for i in range(5,10):
  keys_middle.append(keys[i])
  Values_middle.append(values[i])
for i in range(10,12):
  keys_bottom.append(keys[i])
  Values_bottom.append(values[i])

```

```python

plt.bar(keys_top,Values_top)
plt.title("Mean Earnings per country in relation with players")
plt.xlabel("Name_of_country")
plt.ylabel("Mean_Earning(million dollars)")
plt.grid(True)
plt.show()
```

```python
plt.bar(keys_middle,Values_middle)
plt.title("Mean Earnings per country in relation with players")
plt.xlabel("Name_of_country")
plt.ylabel("Mean_Earning(million dollars)")
plt.grid(True)
plt.show()
```

```python
#Horizontal bar plot
plt.barh(keys_bottom,Values_bottom,color="green")
plt.title("Mean earnings per country in relation with players")
plt.xlabel("Name of country")
plt.ylabel("Mean earnings(million dollars)")
plt.grid(True)
plt.show()
```
