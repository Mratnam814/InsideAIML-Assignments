```python
df[num_col]=df[num_col].fillna(0)
```

```python
df[cat_col]=df[cat_col].fillna('Not_Known')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
for fea in cat_col:
  df[fea]=label_encoder.fit_transform(df[fea])
```

```python
df.head()
```

```python
 from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
 X=df.iloc[:,:-1]
```

```python
X
```

```python
y=df["SalePrice"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
X_train.head()
```

```python
X_train.isnull().sum().sum()
```

```python
from sklearn.feature_selection import mutual_info_regression
```

```python
mutual_info = mutual_info_regression(X_train,y_train)
```

```python
mutual_info=pd.Series(mutual_info)
mutual_info.index = X_train.columns
mutual_info.sort_values(ascending=False)
```

```python
mutual_info.sort_values(ascending=False).plot.bar(figsize=(15,5))
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
Selected_top_columns = SelectKBest(mutual_info_regression,k=20)
```

```python
Selected_top_columns.fit_transform(X_train,y_train)
```

```python
X_train.columns[Selected_top_columns.get_support()]
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_regression = LinearRegression()
```

```python

linear_regression.fit(X_train,y_train)
```

```python
y_pred = linear_regression.predict(X_test)
```

```python
score=r2_score(y_test,y_pred)
```

```python
print(score)
```

```python
MSE= mean_squared_error(y_test,y_pred)
```
