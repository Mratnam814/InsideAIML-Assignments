```python
#10
def is_point_in_triangle(p, p1, p2, p3):
    def sign(p1, p2, p3):
        return (p1[0] - p3[0]) * (p2[1] - p3[1]) - (p2[0] - p3[0]) * (p1[1] - p3[1])

    d1 = sign(p, p1, p2)
    d2 = sign(p, p2, p3)
    d3 = sign(p, p3, p1)

    has_neg = (d1 < 0) or (d2 < 0) or (d3 < 0)
    has_pos = (d1 > 0) or (d2 > 0) or (d3 > 0)

    return not (has_neg and has_pos)

# Example usage:
triangle_vertices = [(9, 3), (6, 8), (3, 6)]
point_to_check = (2, 2)

if is_point_in_triangle(point_to_check, *triangle_vertices):
    print(f"The point {point_to_check} is inside the triangle.")
else:
    print(f"The point {point_to_check} is outside the triangle.")
```

```python
#14
import math

def are_circles_inside(circle1, circle2):
    # circle: (x, y, radius)
    x1, y1, r1 = circle1
    x2, y2, r2 = circle2

    distance_between_centers = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

    # Circle 1 is completely inside Circle 2 if the distance between centers + radius of Circle 1 is less than the radius of Circle 2
    return distance_between_centers + r1 <= r2

# Example usage:
circle1 = (1, 1, 2)  # (x, y, radius) for the first circle
circle2 = (1, 1, 5)  # (x, y, radius) for the second circle

if are_circles_inside(circle1, circle2):
    print("Circle 1 is completely inside Circle 2.")
else:
    print("Circle 1 is not completely inside Circle 2.")
import math

def circumference_of_intersecting_circles(radius1, radius2, distance_between_centers):
    circumference1 = 2 * math.pi * radius1
    circumference2 = 2 * math.pi * radius2

    # Calculate the length of the overlap (common boundary)
    overlap_length = max(0, radius1 + radius2 - distance_between_centers)

    # Calculate the circumference of the intersecting region
    circumference_intersection = circumference1 + circumference2 - overlap_length

    return circumference_intersection

# Example usage:
radius1 = 3
radius2 = 4
distance_between_centers = 5

circumference_intersecting = circumference_of_intersecting_circles(radius1, radius2, distance_between_centers)
print(f"The circumference of the intersecting region is {circumference_intersecting:.2f}")

```

```python
#15
from datetime import datetime, timedelta

def day_of_the_week(year, month, day):
    # Define the base date (Jan. 1, 2016) as a Friday
    base_date = datetime(2016, 1, 1)

    # Calculate the input date
    input_date = datetime(year, month, day)

    # Calculate the difference in days
    days_difference = (input_date - base_date).days

    # Calculate the day of the week
    days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    day_index = (days_difference % 7 + 4) % 7  # +4 to adjust for the base date being a Friday

    return days_of_week[day_index]

# Input date from the user
year = int(input("Enter the year (2016): "))
month = int(input("Enter the month (1-12): "))
day = int(input("Enter the day (1-31): "))

# Check if the date is within the valid range
if 2016 <= year <= 2016 and 1 <= month <= 12 and 1 <= day <= 31:
    day_name = day_of_the_week(year, month, day)
    print(f"The day of the week for {year}/{month}/{day} is {day_name}.")
else:
    print("Invalid date. Please enter a date within the range 2016/1/1 to 2016/12/31.")
```

```python
#7
def find_combinations(n):
    count = 0

    for a in range(10):
        for b in range(10):
            for c in range(10):
                for d in range(10):
                    if a + b + c + d == n:
                        count += 1

    return count

# Input an integer from the user
n = int(input("Enter an integer n: "))

# Find and print the number of combinations
result = find_combinations(n)
print(f"Number of combinations where (a + b + c + d) equals {n}: {result}")
```

```python
#16
def is_rectangle_or_rhombus(side1, side2, diagonal):
    # Check if the parallelogram is a rhombus
    is_rhombus = side1 == side2 and side1**2 + side2**2 == diagonal**2

    if is_rhombus:
        # Check if the rhombus is also a rectangle
        is_rectangle = side1**2 + side2**2 == diagonal**2
        if is_rectangle:
            return "The parallelogram is both a rectangle and a rhombus."
        else:
            return "The parallelogram is a rhombus, but not a rectangle."
    else:
        return "The parallelogram is neither a rectangle nor a rhombus."

# Input side lengths and diagonal from the user
side1 = float(input("Enter the length of side 1: "))
side2 = float(input("Enter the length of side 2: "))
diagonal = float(input("Enter the length of the diagonal: "))

# Check and print the result
result = is_rectangle_or_rhombus(side1, side2, diagonal)
print(result)
```

```python
#17
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def goldbach_conjecture(n):
    if n < 4 or n % 2 != 0:
        print("Please enter an even number greater than or equal to 4.")
        return

    prime_combinations = []

    for i in range(2, n // 2 + 1):
        if is_prime(i) and is_prime(n - i):
            prime_combinations.append((i, n - i))

    print(f"Number of Goldbach combinations for {n}: {len(prime_combinations)}")
    print("Prime combinations:")
    for combo in prime_combinations:
        print(f"{combo[0]} + {combo[1]}")

# Input an even number from the user
while True:
    user_input = int(input("Enter an even number (>=4) to check Goldbach Conjecture: "))
    if user_input >= 4 and user_input % 2 == 0:
        break
    else:
        print("Invalid input. Please enter an even number greater than or equal to 4.")

# Check and print Goldbach combinations
goldbach_conjecture(user_input)
```

```python
#18
def max_number_of_regions(n):
    if n < 0:
        return "Number of lines should be non-negative."
    else:
        return (n * (n + 1) // 2) + 1

# Input the number of lines from the user
while True:
    try:
        n = int(input("Enter the number of lines (non-negative integer): "))
        if n >= 0:
            break
        else:
            print("Please enter a non-negative integer.")
    except ValueError:
        print("Invalid input. Please enter a valid integer.")

# Calculate and print the maximum number of regions
result = max_number_of_regions(n)
print(f"The maximum number of regions obtained by drawing {n} lines is: {result}")
```

```python
#21
def is_palindrome(num):
    # Convert the number to a string for easier comparison
    num_str = str(num)

    # Check if the string representation is equal to its reverse
    return num_str == num_str[::-1]

# Input an integer from the user
try:
    num = int(input("Enter an integer: "))
    if is_palindrome(num):
        print(f"{num} is a palindrome.")
    else:
        print(f"{num} is not a palindrome.")
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```
